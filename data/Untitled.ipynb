```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('bank_transactions_data_2.csv')
```

```python
plt.plot(dataframe["TransactionID"], dataframe['TransactionAmount'] )
```

```python
# Line Plot Univariate
plt.plot(dataframe["TransactionID"], dataframe['TransactionAmount'] )
plt.title("Bank Transactions")
plt.xlabel("TransactionID")
plt.ylabel("Transactions Amount")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dataframe['TransactionAmount'], bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Transaction Amount', fontsize=14)
plt.xlabel('Transaction Amount', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(dataframe['TransactionAmount'], bins=30, color='blue', alpha=0.5, edgecolor='black', label='Transaction Amount')
plt.hist(dataframe['AccountBalance'], bins=30, color='orange', alpha=0.5, edgecolor='black', label='Account Balance')
plt.title('Overlayed Histograms: Transaction Amount and Account Balance', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Violin plot for TransactionAmount
plt.figure(figsize=(10, 6))
sns.violinplot(y='TransactionAmount', data=dataframe, palette='Set2')
plt.title('Violin Plot of Transaction Amount', fontsize=14)
plt.ylabel('Transaction Amount', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Violin plot for TransactionAmount grouped by TransactionType
plt.figure(figsize=(12, 7))
sns.violinplot(x='TransactionType', y='TransactionAmount', data=dataframe, palette='Set3')
plt.title('Violin Plot of Transaction Amount by Transaction Type', fontsize=14)
plt.xlabel('Transaction Type', fontsize=12)
plt.ylabel('Transaction Amount', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Horizontal violin plot for AccountBalance
plt.figure(figsize=(10, 6))
sns.violinplot(x='AccountBalance', data=dataframe, palette='coolwarm')
plt.title('Horizontal Violin Plot of Account Balance', fontsize=14)
plt.xlabel('Account Balance', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
